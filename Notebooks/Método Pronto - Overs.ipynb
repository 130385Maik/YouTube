# Importando as Bibliotecas e o DataFrame

```python
df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/x_FutPythonTrader_Base_de_Dados_Odds_Gols_x.xlsx?raw=true")
```

# Porcentagem de Gols

```python
df['Over05_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 0) else 0, axis=1) 
df['Over15_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 1) else 0, axis=1) 
df['Over25_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 2) else 0, axis=1) 
```

```python
df['Porc_Over05FT_Home'] = (df.groupby('Home')['Over05_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over05FT_Away'] = (df.groupby('Away')['Over05_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Porc_Over15FT_Home'] = (df.groupby('Home')['Over15_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over15FT_Away'] = (df.groupby('Away')['Over15_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Porc_Over25FT_Home'] = (df.groupby('Home')['Over25_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over25FT_Away'] = (df.groupby('Away')['Over25_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
```

# Médias, Desvio Padrão e Coeficiente de Variação

```python
df['Media_Gols_Feitos_Home'] = df.groupby('Home')['FT_Goals_H'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Home'] = df.groupby('Home')['FT_Goals_A'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)

df['Media_Gols_Feitos_Away'] = df.groupby('Away')['FT_Goals_A'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Away'] = df.groupby('Away')['FT_Goals_H'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
```

```python
df['DP_Gols_Feitos_Home'] = df.groupby('Home')['FT_Goals_H'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Home'] = df.groupby('Home')['FT_Goals_A'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)

df['DP_Gols_Feitos_Away'] = df.groupby('Away')['FT_Goals_A'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Away'] = df.groupby('Away')['FT_Goals_H'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
```

```python
df['CV_Gols_Feitos_Home'] = (df['DP_Gols_Feitos_Home'] / df['Media_Gols_Feitos_Home'])
df['CV_Gols_Sofridos_Home'] = (df['DP_Gols_Sofridos_Home'] / df['Media_Gols_Sofridos_Home'])

df['CV_Gols_Feitos_Away'] = (df['DP_Gols_Feitos_Away'] / df['Media_Gols_Feitos_Away'])
df['CV_Gols_Sofridos_Away'] = (df['DP_Gols_Sofridos_Away'] / df['Media_Gols_Sofridos_Away'])
```

# Ajustando o DataFrame

```python
df1 = df[['Home','Away','FT_TotalGoals','Porc_Over05FT_Home','Porc_Over05FT_Away','Media_Gols_Feitos_Home','Media_Gols_Sofridos_Home','Media_Gols_Feitos_Away','Media_Gols_Sofridos_Away','CV_Gols_Feitos_Home','CV_Gols_Sofridos_Home','CV_Gols_Feitos_Away','CV_Gols_Sofridos_Away']]
df1 = df1.dropna()
```

```python
df3 = df[['Home','Away','FT_TotalGoals','Porc_Over25FT_Home','Porc_Over15FT_Away','Media_Gols_Feitos_Home','Media_Gols_Sofridos_Home','Media_Gols_Feitos_Away','Media_Gols_Sofridos_Away','CV_Gols_Feitos_Home','CV_Gols_Sofridos_Home','CV_Gols_Feitos_Away','CV_Gols_Sofridos_Away']]
df3 = df3.dropna()
```

# Filtro Estatístico- Over 0.5 FT

```python
# Selecionando o Filtro Estatítisco
flt1 = ((df1.Porc_Over05FT_Home > 90) & 
        (df1.Porc_Over05FT_Away > 90) & 
        (df1.Media_Gols_Feitos_Home > 1) & 
        (df1.Media_Gols_Sofridos_Home > 1) & 
        (df1.Media_Gols_Feitos_Away > 1) & 
        (df1.Media_Gols_Sofridos_Away > 1) & 
        (df1.CV_Gols_Feitos_Home < 1) & 
        (df1.CV_Gols_Sofridos_Home < 1) & 
        (df1.CV_Gols_Feitos_Away < 1) & 
        (df1.CV_Gols_Sofridos_Away < 1)
       )
df05= df1[flt1];
# Ajustando o Índice
df05.reset_index(inplace=True, drop=True)
df05.index = df05.index.set_names(['Nº'])
df05 = df05.rename(index=lambda x: x + 1)
```

# Backtesting

```python
Odd_Minima = 1.07
Stake = 1
green = Stake * (Odd_Minima - 1) 
red = -Stake

# Over 0.5 FT
df05.loc[((df05['FT_TotalGoals']) >= 1), 'Profit_Over05'] = green
df05.loc[((df05['FT_TotalGoals']) == 0), 'Profit_Over05'] = red
df05['Profit_acu_Over05'] = df05.Profit_Over05.cumsum()
df05.Profit_acu_Over05.plot(title='Over 0.5 FT', xlabel='Quantidade de Apostas', ylabel='Stakes')
profit = round(df05.Profit_acu_Over05.tail(1).item(),2)
ROI = round((df05.Profit_acu_Over05.tail(1)/len(df05)*100).item(),2)
print("Profit:",profit,"stakes em", len(df05),"jogos")
print("ROI:",ROI,"%")
```
