Importando as Bibliotecas

```python
import pandas as pd
```

Coletando as planilhas das ligas secundárias

```python
# Argentina
df1 = pd.read_csv("https://www.football-data.co.uk/new/ARG.csv")
# Austria
df2 = pd.read_csv("https://www.football-data.co.uk/new/AUT.csv")
# Brasil
df3 = pd.read_csv("https://www.football-data.co.uk/new/BRA.csv")
```

Concatenando as Planilhas

```python
df = [df1, df2, df3]
df = pd.concat(df)
```

Visualizando o arquivo gerado

```python
df
```

Selecionando as colunas que serão utilizadas

```python
df = df[['Country','Date','Home','Away','HG','AG','Res','PH','PD','PA']]
```

Renomenando as colunas

```python
df.columns = ['Liga','Data','Mandante','Visitante','Gols Mandante','Gols Visitante','Resultado','Odds Mandante','Odds Empate','Odds Visitante']
```

Visualizando o Resultado Final

```python
df
```

Exportando arquivo inicial para o Excel

```python
df.to_excel("Ligas_Secundárias.xlsx")
```
