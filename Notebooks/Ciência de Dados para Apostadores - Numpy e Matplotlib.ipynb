# Numpy

```python
import numpy as np
```

```python
aposta = ['back_home', 'btts', 'over25', 'back_away', 'over05ht']
odds = [1.80, 2.20, 1.65, 2.40, 1.54]
stake = [100, 100, 100, 100, 100]
```

```python
media_odds = sum(odds)/len(odds)
```

```python
print('Odd Média =',media_odds)
```

```python
# Média com Numpy
media_odds2 = np.mean(odds)
print('Odd Média =',media_odds2)
```

```python
# Desvio Padrao com Numpy
desvio_padrão_odds = np.std(odds)
print('Desvio Padrão das Odds =',desvio_padrão_odds)
```

```python
# Variância com Numpy
variancia_odds = np.var(odds)
print('Variância das Odds =',variancia_odds)
```

```python
# Soma com Numpy 
soma_odds = np.sum(odds)
print('Soma das Odds =',soma_odds)
```

```python
# Soma Acumulada com Numpy 
soma_acu_odds = np.cumsum(odds)
print('Soma Acumulada das Odds =',soma_acu_odds)
```

```python
# Multiplicando Listas com Numpy
lucro = np.dot(odds, stake) - np.sum(stake)
print('O lucro das apostas é =',lucro)
```

```python
# Transformando a Lista em um Array
odds = np.array([1.80, 2.20, 1.65, 2.40, 1.54])
stake = np.array([100, 100, 100, 100, 100])
```
