```python
# Criando um DataFrame a partir de um Dicionário
Lay_Goleada_Punter = {'Mandante' : ['Norrkoping', 'Trelleborgs','IA Akranes'],
                      'Visitante' : ['IFK Goteborg', 'Skovde Aik', 'Fram'],
                      'Valor da Odd' : [9.40, 13.50, 16],
                      'Lucro Sobre a Stake (em %)' : [11.9, 8.0, 6.7]}
```

```python
df_LGP = pd.DataFrame(Lay_Goleada_Punter)
```

```python
df_LGP
```

```python
df_LGP = pd.DataFrame(df_LGP, columns = ['Mandante','Visitante','Lucro Sobre a Stake (em %)','Campeonato'])
```

```python
df_LGP['Campeonato'] = 'Suécia'
```

```python
df_LGP
```
