# Coletando o DataFrame do Football Data

```python
# Ligas Principais
df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/F1.csv")
df['Liga'] = "França 1"
df = df[['Liga','Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR','B365H','B365A']]
df.columns = ['Liga','Date','Home','Away','FT_Goals_H','FT_Goals_A','FT_Result','Odds_H','Odds_A']

# Ligas Secundárias
# df = pd.read_csv("https://www.football-data.co.uk/new/BRA.csv")
# df = df[['League','Season','Date','Home','Away','HG','AG','Res','PH','PA']]
# df.columns = ['Liga','Temporada','Date','Home','Away','FT_Goals_H','FT_Goals_A','FT_Result','Odds_H','Odds_A']
# flt = df.Temporada == 2021
# df = df[flt]

df = df.dropna()
# Ajustando o Índice
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
```

# Criando as Colunas

```python
Stake = 1
Back_Home = Stake * (df.Odds_H - 1)
Back_Away = Stake * (df.Odds_A - 1)
Red = -Stake

df['Back_Home'] = Back_Home.where(df.FT_Result == 'H', other=Red)
df['Back_Away'] = Back_Away.where(df.FT_Result == 'A', other=Red)
```

```python
df
```

# Calculando o ROI do Back Home e Back Away

```python
Back_Home_PL = round(df[['Home','Back_Home']].groupby('Home').sum(),2)
Back_Away_PL = round(df[['Away','Back_Away']].groupby('Away').sum(),2)

ROI_Back_Home = round((Back_Home_PL / len(Back_Home_PL))*100,2)
ROI_Back_Away = round((Back_Away_PL / len(Back_Away_PL))*100,2)

df01 = Back_Home_PL
df02 = Back_Away_PL
df03 = ROI_Back_Home
df04 = ROI_Back_Away

df_back = [df01,df02,df03,df04]
df_back = pd.concat(df_back, axis=1)
df_back.index = df_back.index.set_names(['Time'])
df_back.columns = ['Back_Home_PL','Back_Away_PL','ROI_Back_Home','ROI_Back_Away']
```
