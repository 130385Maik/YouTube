```python
# 1 = Verdadeiro e 0 = Falso
df['Over05_HT'] = df.apply(lambda row: 1 if (row['HT_TotalGoals'] > 0) else 0, axis=1)
df['Over05_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 0) else 0, axis=1)   
df['Over15_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 1) else 0, axis=1)
df['Over25_FT'] = df.apply(lambda row: 1 if (row['FT_TotalGoals'] > 2) else 0, axis=1) 
df['BTTS'] = df.apply(lambda row: 1 if (row['FT_Goals_H'] > 0 and row['FT_Goals_A'] > 0) else 0, axis=1) 
```

```python
# Percentual mercado de gols
df['Media_Over05_HT_Home'] = (df.groupby('Home')['Over05_HT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Media_Over05_HT_Away'] = (df.groupby('Away')['Over05_HT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Media_Over05_FT_Home'] = (df.groupby('Home')['Over05_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Media_Over05_FT_Away'] = (df.groupby('Away')['Over05_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Media_Over15_FT_Home'] = (df.groupby('Home')['Over15_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Media_Over15_FT_Away'] = (df.groupby('Away')['Over15_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Media_Over25_FT_Home'] = (df.groupby('Home')['Over25_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Media_Over25_FT_Away'] = (df.groupby('Away')['Over25_FT'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)

df['Media_BTTS_Home'] = (df.groupby('Home')['BTTS'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
df['Media_BTTS_Away'] = (df.groupby('Away')['BTTS'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True) * 100).shift(1)
```

# Calculando as Médias o Desvio Padrão e o Coeficiente de Variação

```python
# Half-Time (HT)
df['Media_Gols_Feitos_Home_HT'] = df.groupby('Home')['HT_Goals_H'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Home_HT'] = df.groupby('Home')['HT_Goals_A'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Feitos_Away_HT'] = df.groupby('Away')['HT_Goals_A'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Away_HT'] = df.groupby('Away')['HT_Goals_H'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)

df['DP_Gols_Feitos_Home_HT'] = df.groupby('Home')['HT_Goals_H'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Home_HT'] = df.groupby('Home')['HT_Goals_A'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Feitos_Away_HT'] = df.groupby('Away')['HT_Goals_A'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Away_HT'] = df.groupby('Away')['HT_Goals_H'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)

df['CV_Gols_Feitos_Home_HT'] = (df['DP_Gols_Feitos_Home_HT'] / df['Media_Gols_Feitos_Home_HT']).shift(1)
df['CV_Gols_Sofridos_Home_HT'] = (df['DP_Gols_Sofridos_Home_HT'] / df['Media_Gols_Sofridos_Home_HT']).shift(1)
df['CV_Gols_Feitos_Away_HT'] = (df['DP_Gols_Feitos_Away_HT'] / df['Media_Gols_Feitos_Away_HT']).shift(1)
df['CV_Gols_Sofridos_Away_HT'] = (df['DP_Gols_Sofridos_Away_HT'] / df['Media_Gols_Sofridos_Away_HT']).shift(1)
```

```python
# Full-Time (FT)
df['Media_Gols_Feitos_Home_FT'] = df.groupby('Home')['FT_Goals_H'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Home_FT'] = df.groupby('Home')['FT_Goals_A'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Feitos_Away_FT'] = df.groupby('Away')['FT_Goals_A'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Sofridos_Away_FT'] = df.groupby('Away')['FT_Goals_H'].rolling(window=5, min_periods=5).mean().reset_index(0,drop=True).shift(1)

df['DP_Gols_Feitos_Home_FT'] = df.groupby('Home')['FT_Goals_H'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Home_FT'] = df.groupby('Home')['FT_Goals_A'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Feitos_Away_FT'] = df.groupby('Away')['FT_Goals_A'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Sofridos_Away_FT'] = df.groupby('Away')['FT_Goals_H'].rolling(window=5, min_periods=5).std().reset_index(0,drop=True).shift(1)

df['CV_Gols_Feitos_Home_FT'] = (df['DP_Gols_Feitos_Home_FT'] / df['Media_Gols_Feitos_Home_FT']).shift(1)
df['CV_Gols_Sofridos_Home_FT'] = (df['DP_Gols_Sofridos_Home_FT'] / df['Media_Gols_Sofridos_Home_FT']).shift(1)
df['CV_Gols_Feitos_Away_FT'] = (df['DP_Gols_Feitos_Away_FT'] / df['Media_Gols_Feitos_Away_FT']).shift(1)
df['CV_Gols_Sofridos_Away_FT'] = (df['DP_Gols_Sofridos_Away_FT'] / df['Media_Gols_Sofridos_Away_FT']).shift(1)
```

# Critérios de Entrada

```python
# Selecionando os Critérios de Entrada
Over05HT_Home = 0
Over05HT_Away = 0
Over05FT_Home = 0
Over05FT_Away = 0
Over15FT_Home = 0
Over15FT_Away = 0
Over25FT_Home = 70
Over25FT_Away = 70
BTTS_Home = 0
BTTS_Away = 0
GolsTotal_HT_Home = 0
GolsTotal_HT_Away = 0
GolsTotal_FT_Home = 0
GolsTotal_FT_Away = 0
filtro =    (
             (df['Media_Over05_HT_Home'] >= Over05HT_Home) & (df['Media_Over05_HT_Away'] >= Over05HT_Away) &
             (df['Media_Over05_FT_Home'] >= Over05FT_Home) & (df['Media_Over05_FT_Away'] >= Over05FT_Home) &
             (df['Media_Over15_FT_Home'] >= Over15FT_Home) & (df['Media_Over15_FT_Away'] >= Over15FT_Away) &
             (df['Media_Over25_FT_Home'] >= Over25FT_Home) & (df['Media_Over25_FT_Away'] >= Over25FT_Away) &
             (df['Media_BTTS_Home'] >= BTTS_Home) & (df['Media_BTTS_Away'] >= BTTS_Away) &
             ((df['Media_Gols_Feitos_Home_HT'] + df['Media_Gols_Sofridos_Home_HT']) >= GolsTotal_HT_Home) & 
             ((df['Media_Gols_Feitos_Away_HT'] + df['Media_Gols_Sofridos_Away_HT']) >= GolsTotal_HT_Away) &
             ((df['Media_Gols_Feitos_Home_FT'] + df['Media_Gols_Sofridos_Home_FT']) >= GolsTotal_FT_Home) & 
             ((df['Media_Gols_Feitos_Away_FT'] + df['Media_Gols_Sofridos_Away_FT']) >= GolsTotal_FT_Away)
            )

#filtro2 = ((df['CV_Gols_Feitos_Home_FT'] <= 1) & (df['CV_Gols_Sofridos_Home_FT'] <= 1) & (df['CV_Gols_Feitos_Away_FT'] <= 1) & (df['CV_Gols_Sofridos_Away_FT'] <= 1))


df2 = df[filtro]
df2 = df2.dropna()
# Ajustando o Índice
df2.reset_index(inplace=True, drop=True)
df2.index = df2.index.set_names(['Nº'])
df2 = df2.rename(index=lambda x: x + 1)

Stake = 1
green = Stake * (df2.FT_Odds_Under25 - 1) 
red = -Stake

# Over 0.5 HT
df2.loc[((df2['FT_TotalGoals']) < 2), 'Profit_Over15FT'] = green
df2.loc[((df2['FT_TotalGoals']) >= 2), 'Profit_Over15FT'] = red
df2['Profit_acu_Over15FT'] = df2.Profit_Over15FT.cumsum()
df2.Profit_acu_Over15FT.plot(title='Over 1.5 FT', xlabel='Quantidade de Apostas', ylabel='Stakes')
df2['Profit_acu_Over15FT'].tail(1)
```
