```python
# Importando as Bibliotecas
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
import warnings
warnings.filterwarnings('ignore')
```

```python
# DataFrame do Campeonato Brasileiro (2020 e 2021)
df1 = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/FutPythonTrader_Base_de_Dados.xlsx?raw=true")

# DataFrame do Campeonato Brasileiro (2022)
df2 = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/Brazil%20Serie%20A_2022.xlsx?raw=true")

df = [df1,df2]
df = pd.concat(df)

df.to_excel("Brasil.xlsx")
```

```python
# Filtro de Liga
flt = (df.League == 'Brazil Serie A') & (df.FT_Odds_Over25 < 2)
df = df[flt]
# Ajustando o Índice
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
```

```python
# Criando a coluna de resultado (H, D e A) 
df.loc[((df['FT_Goals_H']) > (df['FT_Goals_A'])), 'FT_Result'] = 'H'
df.loc[((df['FT_Goals_H']) == (df['FT_Goals_A'])), 'FT_Result'] = 'D'
df.loc[((df['FT_Goals_H']) < (df['FT_Goals_A'])), 'FT_Result'] = 'A'
```

```python
# Criando a coluna de Pontos (H e A) 
df.loc[((df['FT_Result']) == "H"), 'Pontos_H'] = 3
df.loc[((df['FT_Result']) == "D"), 'Pontos_H'] = 1
df.loc[((df['FT_Result']) == "A"), 'Pontos_H'] = 0

df.loc[((df['FT_Result']) == "H"), 'Pontos_A'] = 0
df.loc[((df['FT_Result']) == "D"), 'Pontos_A'] = 1
df.loc[((df['FT_Result']) == "A"), 'Pontos_A'] = 3
```

```python
Stake = 1
green = Stake * (df.FT_Odds_H - 1) 
red = -Stake

# Back Home
df.loc[((df['FT_Result']) == "H"), 'Profit_BackHome'] = green
df.loc[((df['FT_Result']) != "H"), 'Profit_BackHome'] = red
df['Profit_acu_BackHome'] = df.Profit_BackHome.cumsum()
df.Profit_acu_BackHome.plot(title='Back Home', xlabel='Quantidade de Apostas', ylabel='Stakes')
df['Profit_acu_BackHome'].tail(1)
```

```python
Stake = 1
green = Stake * (df.FT_Odds_D - 1) 
red = -Stake

# Back Draw
df.loc[((df['FT_Result']) == "D"), 'Profit_BackDraw'] = green
df.loc[((df['FT_Result']) != "D"), 'Profit_BackDraw'] = red
df['Profit_acu_BackDraw'] = df.Profit_BackDraw.cumsum()
df.Profit_acu_BackDraw.plot(title='Back Draw', xlabel='Quantidade de Apostas', ylabel='Stakes')
df['Profit_acu_BackDraw'].tail(1)
```

```python
Stake = 1
green = Stake * (df.FT_Odds_A - 1) 
red = -Stake

# Back Draw
df.loc[((df['FT_Result']) == "A"), 'Profit_BackAway'] = green
df.loc[((df['FT_Result']) != "A"), 'Profit_BackAway'] = red
df['Profit_acu_BackAway'] = df.Profit_BackAway.cumsum()
df.Profit_acu_BackAway.plot(title='Back Away', xlabel='Quantidade de Apostas', ylabel='Stakes')
df['Profit_acu_BackAway'].tail(1)
```
