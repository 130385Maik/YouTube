
# BTTS

**Critérios do Método**

a) Calculamos a Porcentagem de Over 0.5 HT

b) Calculamos a Porcentagem de Over 2.5 FT

c) Calculamos a Média de Gols Total no HT do Mandante

d) Calculamos a Média de Gols Total no HT do Visitante

e) Calculamos o Desvio Padrão e o Coeficiente de Variação dessas Médias

f) Definimos nosso Filtro Estatístico:
1. Porcentagem de Over 0.5 HT Total > 75% 
2. Porcentagem de Over 2.5 FT Total > 50%
3. Média de Gols Total HT Home > 1 
4. Média de Gols Total HT Away > 1
5. CV de Gols Total HT Home < 1 
6. CV de Gols Total HT Away < 1

g) Todos os jogos que possuem esses dados entramos com Odd Mínima de 1.54 




# Importando as Bibliotecas e o DataFrame 

```python
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
import warnings
warnings.filterwarnings('ignore')
```

```python
# DataFrame 2021/2022
df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/DataFrame_2021_2022.xlsx?raw=true")

# DataFrame 2020/2021
# df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/DataFrame_2020_2021.xlsx?raw=true")
```

# Porcentagem de Gols

```python
df['Total_Goals_HT'] = df.apply(lambda row: (row['Goals_H_HT'] + row['Goals_A_HT']), axis=1)
df['Total_Goals_FT'] = df.apply(lambda row: (row['Goals_H_FT'] + row['Goals_A_FT']), axis=1) 

df['Over05_HT'] = df.apply(lambda row: 1 if (row['Total_Goals_HT'] > 0) else 0, axis=1) 
df['Over05_FT'] = df.apply(lambda row: 1 if (row['Total_Goals_FT'] > 0) else 0, axis=1) 
df['Over25_FT'] = df.apply(lambda row: 1 if (row['Total_Goals_FT'] > 2) else 0, axis=1)
df['BTTS'] = df.apply(lambda row: 1 if (row['Goals_H_FT'] > 0 and row['Goals_A_FT'] > 0) else 0, axis=1) 
```

```python
df['Porc_Over05HT_Home'] = (df.groupby('Home')['Over05_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over05HT_Away'] = (df.groupby('Away')['Over05_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)

df['Porc_Over05FT_Home'] = (df.groupby('Home')['Over05_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over05FT_Away'] = (df.groupby('Away')['Over05_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)

df['Porc_Over25FT_Home'] = (df.groupby('Home')['Over25_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_Over25FT_Away'] = (df.groupby('Away')['Over25_FT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)

df['Porc_BTTS_Home'] = (df.groupby('Home')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)
df['Porc_BTTS_Away'] = (df.groupby('Away')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True) * 100).shift(1)
```

# Médias, Desvio Padrão e Coeficiente de Variação

```python
df['Media_Gols_Total_HT_Home'] = df.groupby('Home')['Total_Goals_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Total_HT_Away'] = df.groupby('Away')['Total_Goals_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)

df['Media_Gols_Total_FT_Home'] = df.groupby('Home')['Total_Goals_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
df['Media_Gols_Total_FT_Away'] = df.groupby('Away')['Total_Goals_HT'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)

df['Media_BTTS_Home'] = df.groupby('Home')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
df['Media_BTTS_Away'] = df.groupby('Away')['BTTS'].rolling(window=5, min_periods=1).mean().reset_index(0,drop=True).shift(1)
```

```python
df['DP_Gols_Total_HT_Home'] = df.groupby('Home')['Total_Goals_HT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Total_HT_Away'] = df.groupby('Away')['Total_Goals_HT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)

df['DP_Gols_Total_FT_Home'] = df.groupby('Home')['Total_Goals_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
df['DP_Gols_Total_FT_Away'] = df.groupby('Away')['Total_Goals_FT'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)

df['DP_BTTS_Home'] = df.groupby('Home')['BTTS'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
df['DP_BTTS_Away'] = df.groupby('Away')['BTTS'].rolling(window=5, min_periods=1).std().reset_index(0,drop=True).shift(1)
```

```python
df['CV_Gols_Total_HT_Home'] = (df['DP_Gols_Total_HT_Home'] / df['Media_Gols_Total_HT_Home']).shift(1)
df['CV_Gols_Total_HT_Away'] = (df['DP_Gols_Total_HT_Away'] / df['Media_Gols_Total_HT_Away']).shift(1)

df['CV_Gols_Total_FT_Home'] = (df['DP_Gols_Total_FT_Home'] / df['Media_Gols_Total_FT_Home']).shift(1)
df['CV_Gols_Total_FT_Away'] = (df['DP_Gols_Total_FT_Away'] / df['Media_Gols_Total_FT_Away']).shift(1)

df['CV_BTTS_Home'] = (df['DP_BTTS_Home'] / df['Media_BTTS_Home']).shift(1)
df['CV_BTTS_Away'] = (df['DP_BTTS_Away'] / df['Media_BTTS_Away']).shift(1)
```

# Ajustando o DataFrame

```python
df1 = df[['Home','Away','Goals_H_FT','Goals_A_FT','BTTS','Porc_Over05HT_Home','Porc_Over05HT_Away','Porc_Over05FT_Home','Porc_Over05FT_Away','Porc_Over25FT_Home','Porc_Over25FT_Away','Porc_BTTS_Home','Porc_BTTS_Away','Media_Gols_Total_HT_Home','Media_Gols_Total_HT_Away','Media_Gols_Total_FT_Home','Media_Gols_Total_FT_Away','Media_BTTS_Home','Media_BTTS_Away','CV_Gols_Total_HT_Home','CV_Gols_Total_HT_Away','CV_Gols_Total_FT_Home','CV_Gols_Total_FT_Away','CV_BTTS_Home','CV_BTTS_Away']]
df1 = df1.dropna()
```

# Filtro Estatístico

```python
Porcentagem_Over05HT = 50
Porcentagem_Over05FT = 50
Porcentagem_Over25FT = 50
Porcentagem_BTTS = 50
Media_Gols_HT = 1
Media_Gols_FT = 1
Media_BTTS = 1
CV_Gols_HT = 1
CV_Gols_FT = 1
CV_BTTS = 1

# Selecionando o Filtro Estatítisco
flt2 = (df1.Porc_Over25FT_Home > Porcentagem_Over25FT) & (df1.Porc_Over25FT_Away > Porcentagem_Over25FT) & (df1.Porc_BTTS_Home > Porcentagem_BTTS) & (df1.Porc_BTTS_Away > Porcentagem_BTTS) & (df1.Media_Gols_Total_FT_Home > Media_Gols_FT) & (df1.Media_Gols_Total_FT_Away > Media_Gols_FT) & (df1.CV_Gols_Total_FT_Home < CV_Gols_FT) & (df1.CV_Gols_Total_FT_Away < CV_Gols_FT)
df2 = df1[flt2];
# Ajustando o Índice
df2.reset_index(inplace=True, drop=True)
df2.index = df2.index.set_names(['Nº'])
df2 = df2.rename(index=lambda x: x + 1)
```

# Resultados

```python
df2['Odds_BTTS'] = 1.7
Stake = 1
green = Stake * (df2.Odds_BTTS - 1) 
red = -Stake

# Over 0.5 HT
df2.loc[((df2['BTTS']) == 1), 'Profit_BTTS'] = green
df2.loc[((df2['BTTS']) == 0), 'Profit_BTTS'] = red
df2['Profit_acu_BTTS'] = df2.Profit_BTTS.cumsum()
df2.Profit_acu_BTTS.plot(title='BTTS', xlabel='Quantidade de Apostas', ylabel='Stakes')
```
