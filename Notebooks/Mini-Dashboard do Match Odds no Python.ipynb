# Coletando o DataFrame do Football Data

```python
# Ligas Principais
df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/E0.csv")
df['Liga'] = "Inglaterra 1"
df = df[['Liga','Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR','B365H','B365A']]
df.columns = ['Liga','Date','Home','Away','FT_Goals_H','FT_Goals_A','FT_Result','Odds_H','Odds_A']

# Ligas Secundárias
# df = pd.read_csv("https://www.football-data.co.uk/new/BRA.csv")
# df = df[['League','Season','Date','Home','Away','HG','AG','Res','PH','PA']]
# df.columns = ['Liga','Temporada','Date','Home','Away','FT_Goals_H','FT_Goals_A','FT_Result','Odds_H','Odds_A']
# flt = df.Temporada == 2021
# df = df[flt]

df = df.dropna()
# Ajustando o Índice
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
```

# Criando as Colunas

```python
Stake = 1
Back_Home = Stake * (df.Odds_H - 1)
Back_Away = Stake * (df.Odds_A - 1)
Red = -Stake

df['Back_Home'] = Back_Home.where(df.FT_Result == 'H', other=Red)
df['Back_Away'] = Back_Away.where(df.FT_Result == 'A', other=Red)
```

# Criando o Gráfico de times mais lucrativos no Back Home

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
ax = df.groupby(['Home']).sum()['Back_Home'].plot(kind='bar')
plt.ylabel('Stakes')
plt.xlabel('Home Team')
plt.show()
```

# Criando um Mini-Dashboard

```python
wid_home = widgets.Dropdown(name = 'Home')
wid_home.options = list(sorted(df.Home.unique()))
wid_home
```

```python
df_tmp = df[df['Home'] == wid_home.value].copy()
ax = df_tmp.groupby(['Date']).sum()['Back_Home'].plot(kind='bar')
ax.legend()
plt.ylabel('Stakes')
plt.xlabel('Date')
print(wid_home.value)
plt.show()
```

```python
from ipywidgets.widgets.interaction import interactive
wid_home = widgets.Dropdown(name = 'Home')
wid_home.options = list(sorted(df.Home.unique()))

def func(home):

    df_tmp = df[df['Home'] == wid_home.value].copy()
    
    ax = df_tmp.groupby(['Date']).sum()['Back_Home'].plot(kind='bar')
    plt.ylabel('Stakes')
    plt.xlabel('Date')
    plt.show()

interactive_plot = widgets.interactive(func, home=wid_home)

output = interactive_plot.children[-1]
output.layout.height = '350px'

interactive_plot
```
