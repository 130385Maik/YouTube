# Estruturas de Decisão - if, elif, else

```python
#if condição:
#  codigo que vai ser executado se a condição for satisfeita
```

```python
time1 = "São Paulo"
time2 = "Palmeiras"
Odd1 = 4
Odd2 = 2
```

```python
if Odd1 > Odd2:
  print("O ", time2, "tem mais probabilidade de ganhar que o", time1)
elif Odd1 < Odd2:
  print("O ", time1, "tem mais probabilidade de ganhar que o", time2)
else:
  print("O empate terá a maior probabilidade")
```

# Estruturas de Repetição - for() e While()

```python
times = ['São Paulo', 'Chelsea', 'Ceará', 'Floresta', 'Icasa'] 
```

```python
for i in times:
  print(i)
```

```python
i = 10
while i > 0:
  print(i)
  i =  i - 1
```
