# Importando o Dataframe

```python
# Campeonato Brasileiro 2022
df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/Brazil%20Serie%20A_2022.xlsx?raw=true")

# Selecionando as Colunas
df = df[['Home','Away','FT_Odds_H','FT_Odds_D','FT_Odds_A','HT_Odds_Over05','FT_Odds_Over05','FT_Odds_Over15','FT_Odds_Over25','HT_Goals_H','HT_Goals_A','FT_Goals_H','FT_Goals_A','FT_TotalGoals']]
```

# Rodando o Pandas-Profiling

```python
profile = ProfileReport(df,title="Pandas Profiling Report", html={'style':{'full_widht':True}})
```

```python
profile.to_widgets()
```
