# Importando o DataSet

```python
import pandas as pd
df = pd.read_excel("https://www.football-data.co.uk/new/BRA.xlsx")
```

```python
df = df[['Season','Date','Home','Away','HG','AG','Res','PH','PD','PA']]
df.loc[df['Res'] == "H", 'BackHome'] = 1
df.loc[df['Res'] != "H", 'BackHome'] = 0

df = df[['PH','PD','PA','BackHome']]
df
```

# Pycaret

```python
from pycaret.classification import *
s = setup(df, target = 'BackHome')
```

```python
best = compare_models()
```

```python
plot_model(best)
```

```python
evaluate_model(best)
```
