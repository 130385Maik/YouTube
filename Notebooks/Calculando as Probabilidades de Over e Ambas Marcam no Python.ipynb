# Importando as Bibliotecas

```python
# Importando as Bibliotecas
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
import warnings
warnings.filterwarnings('ignore')
```

# Coletando o DataFrame do Football Data

```python
df = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/E0.csv")
df['Liga'] = "Inglaterra 1"
df = df[['Liga','Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR','HTHG','HTAG','HTR']]
df.columns = ['Liga','Date','Home','Away','FT_Goals_H','FT_Goals_A','FT_Result','HT_Goals_H','HT_Goals_A','HT_Result']
df = df.dropna()
# Ajustando o Índice
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
```

# Criando as Colunas de Over Gols e BTTS

```python
# Para Criar a Coluna Total Gols HT e FT
df['HT_TotalGoals'] = df['HT_Goals_H'] + df['HT_Goals_A']
df['FT_TotalGoals'] = df['FT_Goals_H'] + df['FT_Goals_A']
```

```python
# Over 0.5 HT
df.loc[((df['HT_TotalGoals']) == 0), 'Over05HT'] = 0
df.loc[((df['HT_TotalGoals']) >= 1), 'Over05HT'] = 1
df["Over05HT"] = df["Over05HT"].astype(int)

# Over 0.5 FT
df.loc[((df['FT_TotalGoals']) == 0), 'Over05FT'] = 0
df.loc[((df['FT_TotalGoals']) >= 1), 'Over05FT'] = 1
df["Over05FT"] = df["Over05FT"].astype(int)

# Over 1.5 FT
df.loc[((df['FT_TotalGoals']) <= 1), 'Over15FT'] = 0
df.loc[((df['FT_TotalGoals']) >= 2), 'Over15FT'] = 1
df["Over15FT"] = df["Over15FT"].astype(int)

# Over 2.5 FT
df.loc[((df['FT_TotalGoals']) <= 2), 'Over25FT'] = 0
df.loc[((df['FT_TotalGoals']) >= 3), 'Over25FT'] = 1
df["Over25FT"] = df["Over25FT"].astype(int)

# Over 3.5 FT
df.loc[((df['FT_TotalGoals']) <= 3), 'Over35FT'] = 0
df.loc[((df['FT_TotalGoals']) >= 4), 'Over35FT'] = 1
df["Over35FT"] = df["Over35FT"].astype(int)

# Over 4.5 FT
df.loc[((df['FT_TotalGoals']) <= 4), 'Over45FT'] = 0
df.loc[((df['FT_TotalGoals']) >= 5), 'Over45FT'] = 1
df["Over45FT"] = df["Over45FT"].astype(int)

# BTTS (Ambas Marcam)
df['BTTS'] = df.apply(lambda row: 1 if (row['FT_Goals_H'] > 0 and row['FT_Goals_A'] > 0) else 0, axis=1)
```

# Calculando as Médias de Gols e BTTS

```python
media_over05HT_H = round(df[['Home','Over05HT']].groupby('Home').mean() * 100, 2)
media_over05HT_A = round(df[['Away','Over05HT']].groupby('Away').mean() * 100, 2)
media_over05FT_H = round(df[['Home','Over05FT']].groupby('Home').mean() * 100, 2)
media_over05FT_A = round(df[['Away','Over05FT']].groupby('Away').mean() * 100, 2)
media_over15FT_H = round(df[['Home','Over15FT']].groupby('Home').mean() * 100, 2)
media_over15FT_A = round(df[['Away','Over15FT']].groupby('Away').mean() * 100, 2)
media_over25FT_H = round(df[['Home','Over25FT']].groupby('Home').mean() * 100, 2)
media_over25FT_A = round(df[['Away','Over25FT']].groupby('Away').mean() * 100, 2)
media_over35FT_H = round(df[['Home','Over35FT']].groupby('Home').mean() * 100, 2)
media_over35FT_A = round(df[['Away','Over35FT']].groupby('Away').mean() * 100, 2)
media_over45FT_H = round(df[['Home','Over45FT']].groupby('Home').mean() * 100, 2)
media_over45FT_A = round(df[['Away','Over45FT']].groupby('Away').mean() * 100, 2)
media_BTTS_H = round(df[['Home','BTTS']].groupby('Home').mean() * 100, 2)
media_BTTS_A = round(df[['Away','BTTS']].groupby('Away').mean() * 100, 2)

df01 = media_over05HT_H
df02 = media_over05HT_A
df03 = media_over05FT_H
df04 = media_over05FT_A
df05 = media_over15FT_H
df06 = media_over15FT_A
df07 = media_over25FT_H
df08 = media_over25FT_A
df09 = media_over35FT_H
df10 = media_over35FT_A
df11 = media_over45FT_H
df12 = media_over45FT_A
df13 = media_BTTS_H
df14 = media_BTTS_A
df_estatisticas = [df01,df02,df03,df04,df05,df06,df07,df08,df09,df10,df11,df12,df13,df14]
df_estatisticas = pd.concat(df_estatisticas, axis=1)
df_estatisticas.index = df_estatisticas.index.set_names(['Time'])
df_estatisticas.columns = ['Over05HT_Home','Over05HT_Away','Over05FT_Home','Over05FT_Away','Over15FT_Home','Over15FT_Away','Over25FT_Home','Over25FT_Away','Over35FT_Home','Over35FT_Away','Over45FT_Home','Over45FT_Away','BTTS_Home','BTTS_Away']
```

# Apresentando os Resultados

```python
display(df_estatisticas)
print('');print('Média da Liga:');print('')
display(df_estatisticas.mean())
```
