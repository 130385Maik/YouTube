# Como converter Odd em Probabilidade

```python
Odd = float(input("Digite o Valor da Odd: "))
Porcentagem = 100/Odd

print('');print("A Porcentagem correspondente a essa Odd é %.2f" % Porcentagem, "%")
```

# Como converter Probabilidade em Odd

```python
Probabilidade = float(input("Digite o Valor da Probabilidade em %: "))
Odd = 100/Probabilidade

print('');print("A Odd correspondente a essa Porcentagem é %.2f" % Odd)
```

# Calculando a Responsabilidade no Lay

```python
Valor_Apostado_em_Lay = float(input("Digite o Valor da Stake: "))
Odd_do_Lay = float(input("Digite a Odd do Lay: "))

Responsabilidade = Valor_Apostado_em_Lay * (Odd_do_Lay - 1)

print('');print("A sua responsabilidade será R$ %.2f" % Responsabilidade)
```

# Encontrando a Odd do Back Correspondente ao Lay

```python
Odd_do_Lay = float(input("Digite a Odd do Lay: "))
Odd_do_Back = Odd_do_Lay / (Odd_do_Lay - 1)

print('');print("A Odd do Back correspondente a esse Lay é %.2f" % Odd_do_Back)
```

# Como Fechar um Trade em Hedge (Entrada em Back)

```python
Odd_do_Back = float(input("Digite a Odd do Back: "))
Odd_do_Lay = float(input("Digite a Odd do Lay: "))
Valor_Apostado = float(input("Digite o Valor da Stake: "))

Hedge_Back = (Odd_do_Back/Odd_do_Lay) * Valor_Apostado

print('');print("Você deverá apostar R$ %.2f em Lay" % Hedge_Back)
```

# Quanto Vale o Tic - Lay

```python
Odd_do_Lay = float(input("Digite a Odd do Lay: "))
Odd_do_Back = float(input("Digite a Odd do Back: "))
Valor_Apostado_Lay = float(input("Digite o Valor Apostado em Lay: "))

Tic_Lay = (1 - (Odd_do_Lay/Odd_do_Back)) * Valor_Apostado_Lay

print('');print("O Tic nessa Odd vale R$ %.2f centavos" % Tic_Lay)
```

# Rentabilidade de uma Aposta em Back

# Rentabilidade de uma Aposta em Lay

```python
Odd_do_Lay = float(input("Digite a Odd do Lay: "))

Rentabilidade_Lay = (1/(Odd_do_Lay - 1)) * 100

print('');print("Essa aposta vai te pagar %.2f" % Rentabilidade_Lay, "%")
```

# Rentabilidade de um Trade Saindo em Hedge - Back

```python
Odd_do_Back = float(input("Digite a Odd do Back: "))
Odd_do_Lay = float(input("Digite a Odd do Lay: "))

Rentabilidade_Hedge_Back = ((Odd_do_Back/Odd_do_Lay) - 1) * 100

print('');print("Essa aposta vai te pagar %.2f" % Rentabilidade_Hedge_Back, "%")
```

# Rentabilidade de um Trade Saindo em Hedge - Lay


```python
Odd_do_Lay = float(input("Digite a Odd do Lay: "))
Odd_do_Back = float(input("Digite a Odd do Back: "))

Rentabilidade_Hedge_Lay = ((1-(Odd_do_Lay/Odd_do_Back))/(Odd_do_Lay - 1)) * 100

print('');print("Essa aposta vai te pagar %.2f" % Rentabilidade_Hedge_Lay, "%")
```

# Rentabilidade de uma Freebet - Back

```python
Odd_do_Back = float(input("Digite a Odd do Back: "))
Odd_do_Lay = float(input("Digite a Odd do Lay: "))

Rentabilidade_Freebet_Back = (Odd_do_Back-Odd_do_Lay) * 100

print('');print("Essa aposta vai te pagar %.2f" % Rentabilidade_Freebet_Back, "%")
```
