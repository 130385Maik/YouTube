```python
# Importando as Bibliotecas
import pandas as pd
from google.colab import data_table
data_table.enable_dataframe_formatter()
import warnings
warnings.filterwarnings('ignore')
```

```python
# DataFrame do Campeonato Brasileiro
# df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/Brazil%20Serie%20A_2022.xlsx?raw=true")
df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/Brazil%20Serie%20B_2022.xlsx?raw=true")

# Filtro de Odds
flt = (df.FT_Odds_H <= 2)
df = df[flt]
# Ajustando o Índice
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
```

```python
# df.info()
```

```python
df['BTTS'] = df.apply(lambda row: 1 if (row['FT_Goals_H'] > 0 and row['FT_Goals_A'] > 0) else 0, axis=1)  
```

```python
# df.head()
```

```python
Stake = 1
green = Stake * (df.FT_Odds_BTTS_Yes - 1) 
red = -Stake

# BTTS Yes
df.loc[((df['BTTS']) == 1), 'Profit_BTTS_Yes'] = green
df.loc[((df['BTTS']) == 0), 'Profit_BTTS_Yes'] = red
df['Profit_acu_BTTS_Yes'] = df.Profit_BTTS_Yes.cumsum()
df.Profit_acu_BTTS_Yes.plot(title='BTTS Yes', xlabel='Quantidade de Apostas', ylabel='Stakes')
df['Profit_acu_BTTS_Yes'].tail(1)
```

```python
Stake = 1
green = Stake * (df.FT_Odds_BTTS_No - 1) 
red = -Stake

# BTTS No
df.loc[((df['BTTS']) == 0), 'Profit_BTTS_No'] = green
df.loc[((df['BTTS']) == 1), 'Profit_BTTS_No'] = red
df['Profit_acu_BTTS_No'] = df.Profit_BTTS_No.cumsum()
df.Profit_acu_BTTS_No.plot(title='BTTS No', xlabel='Quantidade de Apostas', ylabel='Stakes')
df['Profit_acu_BTTS_No'].tail(1)
```
