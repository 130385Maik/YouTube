Importando as Bibliotecas

```python
import pandas as pd
```

Coletando as planilhas da temporadas passada e atual da liga de sua escolha

```python
liga = input("Digite a Liga Desejada: ")

if liga == "Inglaterra 1":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/E0.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/E0.csv")
elif liga == "Inglaterra 2":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/E1.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/E1.csv")
elif liga == "Alemanha 1":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/D1.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/D1.csv")
elif liga == "Alemanha 2":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/D2.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/D2.csv")
elif liga == "Itália 1":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/I1.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/I1.csv")
elif liga == "Itália 2":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/I2.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/I2.csv")
elif liga == "Espanha 1":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/SP1.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/SP1.csv")
elif liga == "Espanha 2":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/SP2.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/SP2.csv")
elif liga == "França 1":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/F1.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/F1.csv")
elif liga == "França 2":
  df1 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2122/F2.csv")
  df2 = pd.read_csv("https://www.football-data.co.uk/mmz4281/2021/F2.csv")
else:
  print("Você digitou a liga de maneira errada. Tente novamente.")
```

Concatenando as Planilhas

```python
df = [df1, df2]
df = pd.concat(df)
```

Visualizando o arquivo gerado

```python
df
```

Selecionando as colunas que serão utilizadas

```python
df = df[['Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR','B365H','B365D','B365A']]
```

Renomenando as colunas

```python
df.columns = ['Data','Mandante','Visitante','Gols Mandante','Gols Visitante','Resultado','Odds Mandante','Odds Empate','Odds Visitante']
```

Visualizando o Resultado Final
