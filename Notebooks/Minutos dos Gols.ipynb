```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("https://github.com/futpythontrader/YouTube/blob/main/Base_de_Dados/Brazil%20Serie%20A_2022.xlsx?raw=true")
```

```python
df = df[['Home','Away','FT_Goals_H','FT_Goals_A','Goals_H_Minutes','Goals_A_Minutes']]
```

```python
team = "São Paulo"
df = df.groupby(['Home']).get_group(team)
# Resetando o Index
df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
df
```

```python
flt = df.FT_Goals_H != 0
df = df[flt]
df
```

```python
minutos = list(df['Goals_H_Minutes']); minutos
```

```python
minutos_new = []

for i in minutos:
  minutos_new.append([int(j[1:-1]) for j in i[1:-1].replace(' ','').split(',')])

minutos_new
```

```python
df['Goals_H_Minutes'] = minutos_new; df
```

```python
lista = df['Goals_H_Minutes'].sum()
lista.sort()
lista
```

```python
x = [0,15,30,45,60,75,90]
plt.title('Minutos dos Gols', fontsize = 18)
plt.xlabel('Minutos', fontsize = 14)
plt.ylabel('Quantidade de Gols', fontsize = 14)
plt.tick_params(labelsize=12)
plt.xticks(x)
plt.hist(lista, 5, rwidth=0.8)
plt.show()
```
